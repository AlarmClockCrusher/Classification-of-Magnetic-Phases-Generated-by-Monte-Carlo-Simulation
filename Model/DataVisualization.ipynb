The goal is to assume a given Hamiltonian within a 2D system, as described in the poster, use Monte Carlo simulation method to generate spin textures at different temperatures and different magnetic fields.
The simulation results includes the spin components in x, y and z directions. The plot combines the color map of z component and spin orientation in x-y plane. 

**The data is stored in csv files, where the first 3*28*28 columns are the components in each direction of the 28*28 lattice points.**

```python
import pandas as pd
file = pd.read_csv("https://raw.githubusercontent.com/big-data-analytics-physics/bigdata_6820-JasonGuo0/master/finalProject/docs/D0_6~0_6.csv?token=AsRr1_heZ790bemhgrrl4dumQGuMspJpks5cp5eNwA%3D%3D")
print(file.head())
```

Some of the phases are shown below.

```python
import matplotlib.pyplot as plt
import numpy as np

N=28
x, y = np.meshgrid(range(N), range(N))

xspin = file.iloc[249, 0:N**2].values.reshape(N, N)
yspin = file.iloc[249, N**2:2*(N**2)].values.reshape(N, N)
zspin = file.iloc[249, 2*(N**2):3*(N**2)].values.reshape(N, N)

plt.figure(figsize=(4,4))
ax = plt.subplot(1,1,1)
ax.quiver(x, y, xspin, yspin, units='x', pivot='middle', width=0.15, scale=1, scale_units='x')
ax.scatter(x, y, c=zspin, s=30, cmap='rainbow', alpha=0.55)
```
